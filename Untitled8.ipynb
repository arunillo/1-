```python
#4 task
start = int(input("Введите начало диапазона: "))
end = int(input("Введите конец диапазона: "))

found_even = False

for i in range(start, end + 1):
    if i % 2 == 0:
        print(i, end=" ")
        found_even = True

if not found_even:
    print("В этом диапазоне нет чётных чисел.")
else:
    print()  # для перевода строки
```

```python
# 6 task
student = {
    "name": "Aru Muha",
    "age": 20,
    "course": "Информатика",
    "grades": [4, 5, 3, 5, 4]
}

# Выводим имя и курс
print(f"Имя: {student['name']}, Курс: {student['course']}")

# Вычисляем средний балл
average_grade = sum(student['grades']) / len(student['grades'])
print(f"Средний балл: {average_grade:.2f}")

# Добавляем новую оценку
student['grades'].append(5)
print("Обновлённый словарь:", student)
```

```python
#7 task
import math

def calculate_circle_area(radius):
    return math.pi * radius ** 2

def is_positive(number):
    return number > 0

print(f"Площадь круга (r=5): {calculate_circle_area(5):.2f}")
print(f"Число 10 положительное: {is_positive(10)}")
print(f"Число -5 положительное: {is_positive(-5)}")
```

```python
#8 task
def shift_text(text, shift=3):
    result = []
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result.append(chr((ord(char) - base + shift) % 26 + base))
        else:
            result.append(char)
    return ''.join(result)

text = "Hello, World!"
print(f"Исходный: {text}")
print(f"После сдвига: {shift_text(text)}")
```

```python
#9 task
import random

numbers = [random.randint(-10, 10) for _ in range(15)]
positive_numbers = [x for x in numbers if x > 0]
squares = [x**2 for x in numbers]

print("Исходный список:", numbers)
print("Положительные числа:", positive_numbers)
print("Квадраты чисел:", squares)
```
